```python
import pandas as pd
import numpy as np
```

```python
path = '..\\data\\the_movie_lens\\'
```

```python
# Create new rating file
ratings = pd.read_csv(path + 'ratings_small.csv')
ratings
```

```python
# Remove some Nan value in Tmdb in links file
link_large = pd.read_csv('..\\data\\the_movie_lens\\links.csv')
link_small = pd.read_csv('..\\data\\the_movie_lens\\links_small.csv')
links = pd.concat([link_large, link_small])

nan_movie_id = links[links.tmdbId.isna()].movieId.values
nan_movie_id
```

```python
len(nan_movie_id)
```

```python
# Geting movie id don't have metadata 
movie_ids = np.unique(ratings.movieId.to_numpy().astype(int))

nan_metadata = []
movie_metadatas = pd.read_csv(path + 'movies_metadata.csv')
credits = pd.read_csv(path + 'credits.csv')

for movie_id in movie_ids:
    if movie_id not in nan_movie_id:
        if len(links[links.movieId == movie_id].tmdbId.values) > 0 :
            tmdb_id = int(links[links.movieId == movie_id].tmdbId.values[0])

            if str(tmdb_id) not in movie_metadatas.id.values:
                nan_metadata.append(movie_id)

            if tmdb_id not in  credits.id.values:
                nan_metadata.append(movie_id)
                    
```

```python
len(nan_metadata)
```

```python
nan_metadata.extend(nan_movie_id)
```

```python
len(nan_metadata)
```

```python
for i in nan_metadata:
    remove_index = ratings[ratings.movieId == i].index
    ratings.drop(index=remove_index, inplace=True)
```

```python
ratings
```

```python
ratings.to_csv(path + 'new_ratings_small.csv', index=False)
```

```python
# Create a new DataFrame to store the test set
test_df = pd.DataFrame(columns=ratings.columns)
val_df = pd.DataFrame(columns=ratings.columns)

# Iterate through each user and randomly select 10 interactions for the test set
for user_id in ratings['userId'].unique():
    user_data = ratings[ratings['userId'] == user_id]
    test_data = user_data.sample(n=5, random_state=42)  # Randomly select 10 interactions
    test_df = test_df.append(test_data)
    ratings = ratings.drop(test_data.index)  # Remove the selected data for the test set from the train set

```

```python
# Iterate through each user and randomly select 10 interactions for the test set
for user_id in ratings['userId'].unique():
    user_data = ratings[ratings['userId'] == user_id]
    val_data = user_data.sample(n=5, random_state=42)  # Randomly select 10 interactions
    val_df = val_df.append(val_data)
    ratings = ratings.drop(val_data.index)  # Remove the selected data for the test set from the train set

# Save the train and test sets to CSV files
ratings.to_csv(path + 'train_data.csv', index=False)
test_df.to_csv(path + 'test_data.csv', index=False)
val_df.to_csv(path + 'val_data.csv', index=False)
```
